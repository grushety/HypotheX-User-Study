```python
# STEP 1: Upload CSV file from local machine
from google.colab import files
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

print("‚¨ÜÔ∏è Please upload your dataset CSV file")
uploaded = files.upload()

# STEP 2: Load uploaded CSV
filename = next(iter(uploaded))  # get the first uploaded file
df = pd.read_csv(filename)

# STEP 3: Identify class column
# Change 'Class' to the correct column name if needed
class_column = 'Class'

# Check original distribution
print("\nüìä Original Class Distribution:")
original_dist = df[class_column].value_counts()
print(original_dist)

# Optional: plot original class distribution
original_dist.plot(kind='bar', title='Original Class Distribution')
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# STEP 4: Stratified sampling
sample_size = 80
sampled_df, _ = train_test_split(
    df,
    train_size=sample_size,
    stratify=df[class_column],
    random_state=42
)

# Check sampled distribution
print("\n‚úÖ Sampled Class Distribution (150 rows):")
sample_dist = sampled_df[class_column].value_counts()
print(sample_dist)

# Optional: plot sampled class distribution
sample_dist.plot(kind='bar', title='Sampled Class Distribution (150 rows)', color='green')
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# STEP 5: Save and download
output_filename = "sampled_80_stratified.csv"
sampled_df.to_csv(output_filename, index=False)
files.download(output_filename)
print(f"\n‚úÖ File '{output_filename}' is ready for download!")

```
